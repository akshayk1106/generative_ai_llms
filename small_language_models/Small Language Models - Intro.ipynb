# SLMs

* compact, highly efficient versions of the massive large language models
* more accessible and efficient for those with limited resources
* smaller, faster, and more easily customized than LLMs
* have fewer parameters (typically under 10 billion)
* dramatically reduces the computational costs and energy usage
* focus on specific tasks
* trained on smaller datasets
* great for use on devices with limited resources, like smartphones, tablets, or IoT devices
* well suited for on-premise deployments where privacy and data security are very important
* easy to fine-tune: Their smaller size means they can quickly adapt to niche tasks and specialized domains.
* Faster inference: quicker response times, perfect for real-time applications
* SLMs might not be as versatile or deeply understanding as large models, but they handle specific tasks well. For example, an SLM trained to analyze legal texts could do a better job than a general LLM in that area.

# How SLMs Work?

* Next word prediction
* Transformer architecture with attention
* SLMs use techniques like distillation, pruning, and quantization to become smaller, faster, and more efficient.

**Distillation**

1. The larger model (teacher) is trained on a dataset.
2. The smaller model (student) is trained to mimic the outputs of the teacher model.
3. A specialized loss function measures the difference between the outputs of the two models, guiding the student's learning.

There are several methods of knowledge distillation:

1. Response-based: The student model learns to replicate the final output layer of the teacher model, often using "soft targets" for more nuanced information.
2. Feature-based: Focuses on replicating the intermediate layers of the teacher model, helping the student extract similar patterns from data.
3. Relation-based: Trains the student to understand relationships between different parts of the teacher model, emulating complex reasoning processes.

**Pruning**

Pruning is kind of like trimming what’s not needed. During pruning, parts of the model that aren’t as important—like neurons or parameters that don’t contribute much to the overall performance—are removed. This technique helps to shrink the model without significantly impacting its accuracy. However, pruning can be a bit tricky because if you’re too aggressive, you risk cutting out too much and hurting the model’s performance.

1. A large model is fully trained.
2. Parameters that contribute least to the model's performance are identified.
3. These less important parameters are removed, "pruning" the model to a smaller size.

**Quantization**

Quantization involves using fewer bits to store the model's numbers. Normally, a model might use 32-bit numbers, but with this method, those numbers are reduced to 8-bit values, which are much smaller. This makes the model take up less space and allows it to run faster. The best part is that, even though the numbers are less precise, the model still works well with only a small impact on its accuracy.

1. The original model uses high-precision floating-point numbers for its parameters.
2. These parameters are converted to lower-precision representations.
3. This reduction in precision leads to smaller model sizes and faster computation.

# LLMs vs. SLMs


|LLM|SLM|
|---|---|
|Great at handling complex, sophisticated, and general tasks|Best suited for more adequate, simpler, specialized tasks|
|Better accuracy and performance across different tasks|Great at specialized applications and domain-specific tasks|
|Capable of maintaining context over long passages and providing coherent response|May struggle with complex language tasks and long-range context understanding|
|Ex. general-purpose chatbot|Ex. specialized customer service bot focusing on a specific product line|
|Require significant computational power and memory|More economical in terms of resource consumption|
|Often need specialized hardware like GPUs for inference|Can run on standard hardware and even on devices like Raspberry Pi or smartphone|
|Higher operational costs due to resource demand|Shorter training times, making them more accessible for quick deployments|
|Best for cloud environments where there’s plenty of computing power available|better for places with limited resources. They're more efficient for handling smaller tasks|
|Not ideal for on-device AI because it needs a lot of computing power and relies on an internet connection|Perfect for on-device AI, allowing offline use and faster response times.|
|Not a good fit for edge computing because they require a lot of computing power and can be slow to respond|Great for edge computing, where fast responses and efficient use of resources are important. They enable AI in IoT devices, smart homes, and other edge applications|

# Top Open Source LLMs

ChatGPT (Built on GPT-4 model by OpenAI) and Bard (Built on PaLM2 model by Google), as well as many other popular chatbots, have in common that their underlying LLM are proprietary. That means that they are owned by a company and can only be used by customers after buying a license. That license comes with rights, but also with possible restrictions on how to use the LLM, as well as limited information on the mechanisms behind the technology.

**Benefits of Using Open-Source LLMs**

* Enhanced data security and privacy
* Cost savings and reduced vendor dependency
* Code transparency and language model customization: Access to the workings of LLMs, including their source code, architecture, training data, and mechanism for training and inference. Companies using them can customize them for their particular use cases.
* Active community support and fostering innovation
* Addressing the environmental footprint of AI

**Choosing the Right Open-Source LLM for Your Needs**

- What do you want to do? some of them are only released for research purposes.
- Why do you need a LLM? If you can build your idea without needing LLMs, then don’t use them. It’s not mandatory (and you will probably save a lot of money and prevent further resource use).
- How much accuracy do you need? If you need high accuracy, you should opt for bigger LLMs
- How much money do you want to invest? The bigger the model, the more resources will be required to train and operate the model.
- Can you achieve your goals with a pre-trained model? No need to train an LLM from scracth if exisitng pre-trained llm can be used. Out there there are many versions of open-source LLMs trained for a specific use case. If your idea fits in one of these use cases, just for it.
